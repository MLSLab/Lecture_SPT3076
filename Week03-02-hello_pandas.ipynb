Python의 데이터 구조 중 Dictionary 데이터 구조

````python
import pandas as pd
````

````python
mydict = [{'a': 1, 'b': 2, 'c': 3, 'd': 4},
          {'a': 100, 'b': 200, 'c': 300, 'd': 400},
          {'a': 500, 'b': 20, 'c': 80, 'd': 500 },
          {'a': 1000, 'b': 2000, 'c': 3000, 'd': 4000 }]
````

````python
len(mydict)
````

위와 같은 데이터는 dictionary 형태의 구조를 가지고 있습니다.

dictionary는 다음과 같이 정리될 수 있습니다.

* Dictionaries are used to store data values in key:value pairs.

* A dictionary is a collection which is ordered*, changeable and does not allow duplicates.<br/>
: As of Python version 3.7, dictionaries are ordered. In Python 3.6 and earlier, dictionaries are unordered.

## DataFrame 인덱싱 방법

DataFrame 을 인덱싱 하는 방법은 **[ ] 이용한 바로 인덱싱, loc[ ], iloc[ ]**  세 가지 방법

### 1. 바로 indexing 하기

[ ] 는indexing operator 라고도 부름
(This function also known as indexing operator)

즉, 아래처럼 column을 바로 인덱싱할 수 있다. 
```
df["column"] 
```


````python
df['a']
````

* 두개 이상의 column을 가져올 때? 
```
df[list]
```

````python
new_df = df[['a', 'c']]
````

````python
new_df.shape
````

### 2. df.loc[ "row" , "column" ] 
This function is used for labels  --> Straightforward

````python
df
````

````python
df.loc['tennis', 'b']
````

````python
df.loc[ ['tennis', 'pingpong'], 'b']
````

````python
df.loc[ ['football', 'tennis'], ['c', 'b']]
````

````python
df.loc[ 'football' : 'tennis', 'b': 'd']
````

### 3. iloc 
iloc: Purely integer-location based indexing for selection by position

````python
df
````

````python
df.iloc[1:3]
````

````python
df.iloc[0:3, [0, 3]]
````

````python
df.iloc[:3]
````

````python
df.iloc[2:]
````

````python
df.iloc[2:, 1:3 ]
````

````python
df.iloc[2:, [False, True, False, True]]
````
