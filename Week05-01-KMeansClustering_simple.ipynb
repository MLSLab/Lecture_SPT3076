```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt

#from sklearn.datasets.samples_generator import make_blobs
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
centers = 5
X, y = make_blobs(n_samples=300, centers=centers, cluster_std=0.6, random_state=0)
```

```python
X.shape, y.shape
```

```python
X[:3]
```

```python
plt.plot(X[:, 0], X[:, 1], '.')
```

```python
k = 5
kmeans = KMeans(n_clusters = k)
```

```python
kmeans.fit(X)
```

```python
wcss = kmeans.inertia_
centers = kmeans.cluster_centers_
print(centers.shape)
```

```python
plt.plot(X[:, 0], X[:, 1], '.')
plt.plot(centers[:, 0], centers[:, 1], '*', markersize=15)
plt.title('k = {} / SSE = {:.2f}'.format(k, wcss))
```

```python
v_wcss = [] #empty list
v_k =[]
for i in range(10):
  k = i+1
  #print(k)
  kmeans = KMeans(n_clusters = k)
  kmeans.fit(X)

  wcss = kmeans.inertia_
  centers = kmeans.cluster_centers_

  v_wcss.append(wcss)
  v_k.append(k)
  continue
  plt.figure()
  plt.plot(X[:, 0], X[:, 1], '.')
  plt.plot(centers[:, 0], centers[:, 1], '*', markersize=15)
  plt.title('k = {} / SSE = {:.2f}'.format(k, wcss))
```

```python
v_wcss
```

```python
plt.plot(v_k, v_wcss)
plt.xlabel('K')
plt.ylabel('WCSS')
plt.title('centers={}'.format(centers))
```
