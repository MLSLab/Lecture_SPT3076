Kernel Density Estimation (KDE, 커널밀도추정)

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
total_bill = tips['total_bill']
tip = tips['tip']
```

```python
sns.histplot(total_bill, kde=True)
```

```python
sns.kdeplot(total_bill)
sns.rugplot(total_bill)
```

```python
sns.kdeplot(tip)
sns.rugplot(tip)
```

```python
sns.kdeplot(total_bill, tip)
sns.rugplot(total_bill, y = tip)
```
