# SVM example
Age vs loan 풀어보기

### Data 설정하기

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
age = [25, 35, 45, 20, 35, 52, 23, 40, 60, 48, 33]
```

```python
loan = [40000, 60000, 80000, 20000, 120000, 18000, 95000, 62000, 100000, 220000, 150000]
```

```python
#loan = np.array(loan) / 1000.0 # 숫자가 크므로 1000 으로 나누어 계산
```

```python
default = ['N', 'N', 'N', 'N', 'N', 'N', 'Y', 'Y', 'Y', 'Y', 'Y']
```

```python
# Import LabelEncoder
from sklearn import preprocessing
#creating labelEncoder
le = preprocessing.LabelEncoder()
# Converting string labels into numbers.
default_encoded=le.fit_transform(default)
```

```python
print(default_encoded)
```

### Data 그려보기

```python
fig = plt.figure(figsize=(8, 5), dpi=150)
#ax = fig.add_subplot(111, projection='3d') # Axe3D object

ax = plt.axes(projection='3d')
ax.scatter(age, loan, default_encoded,  c = default_encoded, s= 20, alpha=1.0, marker='o', cmap=plt.cm.rainbow)
ax.set_xlabel('Age')
ax.set_ylabel('Loan')
ax.set_zlabel('Default')
#ax.view_init(10, 135)
```

```python
#combinig weather and temp into single listof tuples
features=list(zip(age, loan))
```

### 기계학습 모델 (분류기, classifier) 설정

```python
from sklearn.svm import SVC
model = SVC(kernel='linear') #classifier
```

### 학습하기 - 데이터 사용

```python
# Train the model using the training sets
model.fit(features, default_encoded)
```

### Test 해보기

```python
model.predict([[48,142000]]) 
```
