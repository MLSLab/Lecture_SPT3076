```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

* Support Vector Machine 과 동일한 예제. 약간 수정하며 고찰

```python
#X = np.array([[1, 2], [1, 5], [2, 2], [3, 3.5], [7, 6], [9, 4], [8, 2]]) 
X = np.array([[1, 2], [1, 5], [2, 2], [3, 3.5], [3, 6], [9, 4], [8, 2]]) 
Y = np.array([1, 1, 1, 1, 0, 0, 0])
```

```python
x1 = X[:, 0]
x2 = X[:, 1]
```

```python
#plt.figure(dpi=150)
plt.scatter(x1, x2, c=Y)
plt.grid('on')
plt.xlabel('x1')
plt.ylabel('x2')
plt.axis('equal')
```

```python
from sklearn import tree
#clf = tree.DecisionTreeClassifier(criterion='entropy')
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y) #Training, 학습과정
```

```python
clf.predict([[2., 4.]])
```

```python
clf.predict([[4., 5.]])
```

```python
from sklearn.tree import export_graphviz
import pydotplus
from IPython.display import Image
dot_data = export_graphviz(clf, out_file=None, feature_names=['x1', 'x2'],
                           filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
img = Image(graph.create_png())
```

```python
img
```

```python
graph.write_png("images/decision_tree_simple_ex3.png")
```

```python
graph.write_svg("images/decision_tree_simple_ex3.svg")
```

```python
from dtreeviz.trees import dtreeviz # remember to load the package

viz = dtreeviz(clf, X, Y,
                target_name="target")
```
